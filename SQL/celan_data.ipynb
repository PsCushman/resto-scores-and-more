```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('resources/Restaurant_Scores_-_LIVES_Standard.csv')
```

```python
df.head()
```

```python
df = df.dropna(subset=['inspection_score'])
```

```python
df['inspection_date'] = pd.to_datetime(df['inspection_date'])

# Convert the 'inspection_date' column to the desired format
df['inspection_date'] = df['inspection_date'].dt.strftime('%Y-%m-%d')
```

```python
df['inspection_score'] = df['inspection_score'].astype(int)
```

```python
df.dtypes
```

```python
useful_df = df[['business_id', 'business_name', 'business_address', 'business_city',
                'business_latitude', 'business_longitude', 'business_location',
                'inspection_id', 'inspection_date',
                'inspection_score', 'violation_id', 'violation_description',
                'risk_category']]
```

```python
useful_df.head()
```

```python
business_info_df = df[['business_id', 'business_name', 'business_address', 'business_city',
                  'business_latitude', 'business_longitude', 'business_location']]
```

```python
business_info_df.head()
```

```python
inspection_info_df = df[['business_id', 'inspection_id', 'inspection_date',
                         'inspection_score', 'violation_id', 'violation_description',
                         'risk_category']]
```

```python
inspection_info_df.head()
```

```python
for_maps_df = df[[ 'business_name', 'business_address', 
                  'business_latitude', 'business_longitude', 'business_location',
                  'inspection_date', 'inspection_score', 'violation_id', 
                  'violation_description','risk_category']]
```

```python
for_maps_df.head()
```

```python
for_maps_df = for_maps_df.dropna(subset=['business_location'])
```

```python
for_maps_df.head()
```

```python
useful_df.to_csv('resources/useful.csv', index=False)
```

```python
business_info_df.to_csv('resources/business_info.csv', index=False)
```

```python
inspection_info_df.to_csv('resources/inspection_info.csv', index=False)
```

```python
for_maps_df.to_csv('resources/for_maps.csv', index=False)
```
